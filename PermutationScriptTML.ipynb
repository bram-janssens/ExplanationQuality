```python
import multiprocessing
import ParallelCharlie
import os
import pandas as pd
import pickle

os.chdir('C:\\Users\\banjanss\\Documents\\2022_StabilityXAI\\Files_Lisa')
files = os.listdir()
files = [each_string.lower() for each_string in files]
files = [el for el in files if 'charlie' in el]
files = [el for el in files if 'train' in el]
files

appended_data = []
for infile in files:
    data = pd.read_csv(infile)
    appended_data.append(data.reset_index(drop=True))
# see pd.concat documentation for more info
appended_data = pd.concat(appended_data)
appended_data = appended_data.drop_duplicates(subset='Unnamed: 0')
appended_data = appended_data.reset_index(drop=True)

if __name__ == '__main__':
    with multiprocessing.Pool(8) as p:
        list_of_permutations = p.map(ParallelCharlie.create_parallel_permutations_tml, range(appended_data.shape[0]))

keys = []
for dataf in list_of_permutations:
    keys.append(dataf['id'][0])
zipbObj = zip(keys, list_of_permutations)
save_dict = dict(zipbObj)

with open('charlie_permuted.pickle', 'wb') as f:
    pickle.dump(save_dict, f)        
```

```python
import multiprocessing
import ParallelCovid
import os
import pandas as pd
import pickle
import numpy as np

os.chdir('C:\\Users\\banjanss\\Documents\\2022_StabilityXAI\\Files_Lisa')
files = os.listdir()
files = [each_string.lower() for each_string in files]
files = [el for el in files if 'covid' in el]
files = [el for el in files if 'train' in el]
files

appended_data = []
for infile in files:
    data = pd.read_csv(infile)
    appended_data.append(data.reset_index(drop=True))
# see pd.concat documentation for more info
appended_data = pd.concat(appended_data)
appended_data = appended_data.drop_duplicates(subset='Unnamed: 0')
appended_data = appended_data.reset_index(drop=True)
appended_data = appended_data[appended_data['Unnamed: 0']!=1629]
appended_data['cleaned_text'] = np.where(appended_data['Unnamed: 0']==1629, 'empty text with url', appended_data['cleaned_text'])

if __name__ == '__main__':
    with multiprocessing.Pool(8) as p:
        list_of_permutations = p.map(ParallelCovid.create_parallel_permutations_tml, appended_data.index)

keys = []
for dataf in list_of_permutations:
    keys.append(dataf['id'][0])
zipbObj = zip(keys, list_of_permutations)
save_dict = dict(zipbObj)

with open('covid_permuted.pickle', 'wb') as f:
    pickle.dump(save_dict, f)        
```
